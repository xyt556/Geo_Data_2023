```python
import os
import folium

print(folium.__version__)
```

```python
m = folium.Map(location=[41, -71], zoom_start=4)

kw = dict(prefix='fa', color='green', icon='arrow-up')

icon = folium.Icon(angle=180, **kw)
folium.Marker([41, -72], icon=icon).add_to(m)

icon = folium.Icon(angle=45, **kw)
folium.Marker([41, -75], icon=icon).add_to(m)

icon = folium.Icon(angle=90, **kw)
folium.Marker([41, -78], icon=icon).add_to(m)

m.save(os.path.join('results', 'Rotate_icon.html'))

m
```
