```python
import os
import folium

print(folium.__version__)
```

## Defaults

```python
m = folium.Map()

folium.Marker(location=[0, 0]).add_to(m)

m.save(os.path.join('results', 'ContinuousWorld_0.html'))

m
```

## World copy jump

```python
m = folium.Map(world_copy_jump=True)

folium.Marker(location=[0, 0]).add_to(m)

m.save(os.path.join('results', 'ContinuousWorld_1.html'))

m
```

## Continuous world

```python
m = folium.Map(continuous_world=True)

folium.Marker(location=[0, 0]).add_to(m)

m.save(os.path.join('results', 'ContinuousWorld_2.html'))

m
```

## No wrap

```python
m = folium.Map(no_wrap=False)

folium.Marker(location=[0, 0]).add_to(m)

m.save(os.path.join('results', 'ContinuousWorld_3.html'))

m
```
